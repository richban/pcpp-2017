```python
import pandas as pa
from bokeh.plotting import figure, output_file, show
from bokeh.io import output_notebook
output_notebook()
```

```python
path_linkedlist = '../LinkedListStack/results.txt'
```

```python
converters = {
    1: lambda x: int(x),
    2: lambda x: float(x),
    3: lambda x: str(x),
    4: lambda x: float(x),
    5: lambda x: int(x),
}
```

```python
columns = ["implementation", "threadCount", "mean", "scale", "standart_deviation", "count"]
```

```python
linked_list = pa.read_csv(path_linkedlist, delim_whitespace=True, index_col=False,
                         header=None, names=columns, converters=converters, skiprows=5)
```

```python
stack = linked_list[:32]
striped_stack = linked_list[32:]
```

```python
# Plot Concurent Synchronized Stack vs Concurrent Striped Stack
p1 = figure(
    title="Concurent Synchronized Stack vs Concurrent Striped Stack",
    x_axis_label='# threads', y_axis_label='time ms'
)

# renders
p1.line(stack['threadCount'], stack['mean'], legend='Synchronized Stack',
        line_color="red", line_width=3)
p1.line(striped_stack['threadCount'], striped_stack['mean'],
        legend="Striped Stack", line_width=3)
```

```python
# output to static HTML file
output_file("SynchronizedStack.StripedStack.html")
show(p1)
```
